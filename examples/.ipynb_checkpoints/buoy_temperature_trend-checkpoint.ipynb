```python
from importlib import reload  # Python 3.4+
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
from ndbc_api import NdbcApi
import concurrent.futures
from tqdm.contrib.concurrent import process_map, thread_map
from tqdm.autonotebook import tqdm
import matplotlib
import seaborn as sns
import warnings
warnings.filterwarnings(
    action="ignore",
    category=UserWarning)
import xbuoy
```

```python
extrema = lambda x: np.array((x.min().values, x.max().values))

buoy_ds = xr.open_dataset("../data/NDBC_CC_Daily.nc")
# buoy_ds = buoy_ds.resample(time="1M").mean("time")
buoy_ds["WTMP"] = buoy_ds["WTMP"] - buoy_ds["WTMP"].mean("time")
```

```python
wtmp_extrema = extrema(buoy_ds["WTMP"]).max().round() / 1.5
buoy_ds = buoy_ds.where(buoy_ds["wtemp_coverage"] > 90)
buoy_ds.coords["time"] = buoy_ds.time.dt.year.values
buoy_ds.coords["time"] = (buoy_ds.coords["time"] - buoy_ds.coords["time"].mean("time")).values
buoy_ds["WTMP_loc_avg"] = buoy_ds["WTMP"].mean("station_id")
buoy_ds["WTMP_loc_spread"] = buoy_ds["WTMP"].std("station_id")
```

```python
p = buoy_ds["WTMP"].polyfit(dim = "time", deg = 1)
fit = xr.polyval(buoy_ds['time'], p.polyfit_coefficients)
```

```python
trends = 10 * buoy_ds["WTMP"].polyfit(dim = "time", deg = 1)["polyfit_coefficients"].sel(degree = 1)
trends["latitude"] = buoy_ds["latitude"]
trends["longitude"] = buoy_ds["longitude"]

trend_max = extrema(trends).max()
```

```python
import cmocean.cm as cm
import cartopy.feature as cfeature

fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(1, 2, 2, projection=ccrs.PlateCarree())
    # Plot each station's location as a scatter point
for station in buoy_ds.station_id:
    station_data = trends.sel(station_id = station)
    if ~np.isnan(station_data.sum()):
        cb = ax.scatter(station_data.longitude, station_data.latitude, c = station_data, 
                   vmin = -trend_max, vmax = trend_max,
                   edgecolors='black', cmap = cm.balance,
                   transform=ccrs.PlateCarree(), s = 70)

# ax.set_title(str(formatted_date));
fig.tight_layout()

# Add coastlines to the map with some transparency
ax.coastlines(alpha=0.5)
ax.add_feature(cfeature.LAND, color = "#d3d3d3")

# Add gridlines with labels on the map, disabling top and right labels
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, alpha=0.1, color = "grey")
gl.top_labels = False
gl.right_labels = False


fig.colorbar(cb, label = "Temperature Trend [deg C / year]", extend = "both", ax = ax, orientation = "horizontal", fraction = 0.035)

ax.set_aspect(1.5)
```
