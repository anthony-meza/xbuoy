```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, RBF, WhiteKernel, ConstantKernel as C

np.random.seed(42)  # For reproducibility

# Define latitude, longitude, and time coordinates
n_days = 200  # Total time points
grid_size = 25

lat_coords = np.linspace(-10, 10, grid_size)  # Latitude range
lon_coords = np.linspace(-10, 10, grid_size)  # Longitude range
time_feature = np.arange(1, n_days + 1)  # Time

# Function to compute a spatially heterogeneous synthetic field with lat, lon, and time
def synthetic_space_time_function(lat, lon, t):
    # Propagating pulse with spatial variation
    pulse_center_lat = 0.0 + 0.05 * t  # Moving pulse in latitude
    pulse_center_lon = 0.0
    pulse_width = 5.0
    spatial_distance = np.sqrt((lat - pulse_center_lat) ** 2 + (lon - pulse_center_lon) ** 2)
    pulse = 15 * np.exp(-(spatial_distance ** 2) / (2 * pulse_width ** 2)) * np.sin(2 * np.pi * t / 30)
    
    # Heterogeneity: add spatial variations (e.g., linear trends in latitude/longitude)
    spatial_heterogeneity = (0.1 * lat + 0.3 * lon)  # Spatial trend
    
    # Seasonal variability (constant across space)
    seasonal_variability = 10 * np.sin(2 * np.pi * t / (365.25 * 0.2))
    
    # Interannual variability (constant across space)
    interannual_variability = 5 * np.sin(2 * np.pi * t / (365.25 * 5))
    
    # Add linear temporal trend
    trend = 0.05 * t
    
    return 20 + trend + seasonal_variability + interannual_variability + pulse + spatial_heterogeneity

# Create synthetic data over lat/lon/time
space_time_grid = xr.DataArray(
    np.zeros((n_days, len(lat_coords), len(lon_coords))),
    coords=[time_feature, lat_coords, lon_coords],
    dims=["time", "lat", "lon"]
)

# Fill the synthetic data array
for t in range(n_days):
    for i, lat in enumerate(lat_coords):
        for j, lon in enumerate(lon_coords):
            space_time_grid[t, i, j] = synthetic_space_time_function(lat, lon, time_feature[t])

# Visualize a snapshot of the synthetic spatiotemporal field
plt.figure(figsize=(8, 6))
space_time_grid.mean("time").plot(cmap="coolwarm")
plt.title("Snapshot of the True Spatiotemporal Field (With Heterogeneity) at t=60")
plt.show()
```

```python
# Parameters
n_buoys = 15  # Number of buoys to generate
lat_range = (lat_coords[0], lat_coords[-1])  # Latitude range
lon_range = (lon_coords[0], lon_coords[-1])  # Longitude range

# Randomly generate buoy locations within the specified latitude and longitude range
random_buoy_locs = np.column_stack((
    np.random.random_integers(low=lat_range[0], high=lat_range[1], size=n_buoys),
    np.random.random_integers(low=lon_range[0], high=lon_range[1], size=n_buoys)
))

# Extract buoy data from the true spatiotemporal field (lat, lon, and time)
buoy_data = xr.DataArray(
    np.zeros((n_days, n_buoys)),
    coords=[time_feature, np.arange(n_buoys)],
    dims=["time", "buoy"]
)

# Add noise to the buoy data
noise_scale = 0.5
for b, (lat, lon) in enumerate(random_buoy_locs):
    buoy_data[:, b] = space_time_grid.sel(lat=lat, lon=lon, method="nearest").values + np.random.normal(0, noise_scale, n_days)

# Calculate the anomalies for each buoy based on the noisy/incomplete data
mean_buoy_data = buoy_data.mean(dim='time')  # Mean of the noisy/incomplete data
buoy_anomalies = buoy_data - mean_buoy_data

# Simulate missing data in the target buoy (during the pulse period)
target_buoy = 0
missing_time_start = 50
missing_time_end = 100  # Large gap
buoy_anomalies[missing_time_start:missing_time_end, target_buoy] = np.nan

# Plot the target buoy data with the simulated gap
plt.figure(figsize=(10, 6))
plt.plot(time_feature, buoy_anomalies[:, target_buoy], label="Target Buoy (Anomalies, With Gaps)", color='blue')
plt.axvspan(missing_time_start, missing_time_end, color='red', alpha=0.3, label="Missing data")
plt.title("Target Buoy Anomalies with a Large Gap")
plt.xlabel("Time (Days)")
plt.ylabel("Temperature Anomalies")
plt.legend()
plt.show()
```

```python
# Prepare training data using anomalies from all buoys (excluding missing data from target buoy)
X_train_full = []
y_train_full = []

# Loop over all buoys to gather training data for GPR
for b in range(len(random_buoy_locs)):
    lat_np = np.full((n_days, 1), random_buoy_locs[b, 0])  # Latitude for each time point
    lon_np = np.full((n_days, 1), random_buoy_locs[b, 1])  # Longitude for each time point
    time_np = time_feature.reshape(-1, 1)  # Time as a feature
    
    X_train_buoy = np.hstack([lat_np, lon_np, time_np])
    y_train_buoy = buoy_anomalies[:, b].values
    
    # If this is the target buoy, exclude the missing data
    if b == target_buoy:
        observed_indices = ~np.isnan(buoy_anomalies[:, target_buoy])
        X_train_full.append(X_train_buoy[observed_indices])
        y_train_full.append(y_train_buoy[observed_indices])
    else:
        # Include all data from other buoys
        X_train_full.append(X_train_buoy)
        y_train_full.append(y_train_buoy)

# Convert lists to numpy arrays
X_train_full = np.vstack(X_train_full)  # Stack all buoy data vertically
y_train_full = np.concatenate(y_train_full)  # Concatenate target values
```

```python
# Initialize the Gaussian Process Regressor with the kernel
# kernel = C(1.0, (1e-3, 1e3)) * RBF(length_scale=[1.0, 1.0, 1.0], length_scale_bounds=(1e-2, 1e2)) + WhiteKernel(noise_level=0.2, noise_level_bounds=(1e-5, 10))
kernel = C(1.0, (1e-3, 1e3)) * Matern(length_scale=[1.0, 1.0, 1.0], length_scale_bounds=(1e-2, 1e2), nu=2.5) + WhiteKernel(noise_level=0.2, noise_level_bounds=(1e-5, 1))

gp_full = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=5, alpha=0, normalize_y=False)

# Train the GPR model using the anomalies dataset
gp_full.fit(X_train_full, y_train_full)
```

```python
print("Initial kernel parameters:")
print(gp_full.kernel.get_params())

print("\nOptimized kernel parameters:")
print(gp_full.kernel_)
```

```python
# Predict the anomalies for the target buoy
lat_target = np.full((n_days, 1), random_buoy_locs[target_buoy, 0])  # Latitude of target buoy
lon_target = np.full((n_days, 1), random_buoy_locs[target_buoy, 1])  # Longitude of target buoy
X_target = np.hstack([lat_target, lon_target, time_feature.reshape(-1, 1)])  # Features for the target buoy

y_pred_full, y_std_full = gp_full.predict(X_target, return_std=True)
```

```python
# Plot the comparison between original data, noisy buoy data (with gap), and GPR predictions (anomalies)
plt.figure(figsize=(10, 6))

# Plot the original data (without noise, anomalies)
plt.plot(time_feature, space_time_grid.sel(lat=random_buoy_locs[target_buoy, 0], lon=random_buoy_locs[target_buoy, 1], method="nearest") - mean_buoy_data[target_buoy], label="Original Data (Anomalies)", color='green')


# Plot the GPR predictions (anomalies)
plt.plot(time_feature, y_pred_full, label="GPR Predictions (Anomalies)", color='orange')

# Plot the uncertainty (1.96 * standard deviation)
plt.fill_between(time_feature, y_pred_full - 1.96 * y_std_full, y_pred_full + 1.96 * y_std_full, color='orange', alpha=0.2, label="95% Confidence Interval")

# Plot the noisy buoy data (with gaps)
plt.plot(time_feature, buoy_anomalies[:, target_buoy], label="Noisy Sampled Data (Anomalies, With Gaps)", color='blue', linestyle='--')

# Highlight the gap
plt.axvspan(missing_time_start, missing_time_end, color='red', alpha=0.3, label="Missing Data Gap")

# Labels and title
plt.xlabel('Time (Days)')
plt.ylabel('Temperature Anomalies')
plt.title('Comparison: Original Anomalies, Noisy Data (With Gap), and GPR Reconstruction (Target Buoy)')
plt.legend()
plt.show()


# Plot the nearby stations' anomalies used to fill the gaps in the target buoy
plt.figure(figsize=(10, 6))

# Plot the time series for each nearby buoy (anomalies)
for i in range(1, len(random_buoy_locs)):
    plt.plot(time_feature, buoy_anomalies[:, i], label=f'Buoy {i} Anomalies')

# Labels and title
plt.xlabel('Time (Days)')
plt.ylabel('Temperature Anomalies')
plt.title('Nearby Buoy Anomalies Used to Fill Target Buoy Gaps')
plt.legend()
plt.show()
```

```python
# Interpolated Spatial Anomaly Plot Using GPR at the Pulse Midpoint
pulse_midpoint = 50

# Create a meshgrid for lat/lon
lat_min, lat_max = min(lat_coords), max(lat_coords)
lon_min, lon_max = min(lon_coords), max(lon_coords)
grid_lat, grid_lon = np.meshgrid(np.linspace(lat_min, lat_max, grid_size), np.linspace(lon_min, lon_max, grid_size))

# Flatten the grid coordinates to create the feature matrix for GPR prediction
grid_points = np.vstack([grid_lat.ravel(), grid_lon.ravel()]).T  # Shape: (grid_size^2, 2)
time_at_pulse = np.full((grid_points.shape[0], 1), pulse_midpoint)  # Shape: (grid_size^2, 1)

# Combine the time and spatial features for the grid
X_grid = np.hstack([grid_points, time_at_pulse])

# Predict the temperature anomaly at each grid point using GPR
y_grid_pred, y_grid_std = gp_full.predict(X_grid, return_std=True)

# Reshape the predictions back to the grid shape for plotting
y_grid_pred_reshaped = y_grid_pred.reshape(grid_size, grid_size)

# Plot the spatial anomaly field (with contours and scatter plot)
plt.figure(figsize=(10, 8))

# Find symmetric color limits
print(np.abs(buoy_anomalies.sel(time = pulse_midpoint)))
vmax = max(np.nanmax(np.abs(buoy_anomalies.sel(time = pulse_midpoint))), np.nanmax(np.abs(y_grid_pred_reshaped)))
vmax = vmax * 1.2
vmin = -vmax

# Create a contour plot for the interpolated temperature anomaly field
contour = plt.pcolormesh(grid_lat, grid_lon, y_grid_pred_reshaped, cmap='coolwarm', vmin=vmin, vmax=vmax)

# Add color bar to show the temperature anomaly scale
plt.colorbar(contour, label='Temperature Anomaly (Predicted)')

# Scatter plot of the buoy locations with their anomalies
plt.scatter(random_buoy_locs[:, 0], random_buoy_locs[:, 1], c=buoy_anomalies.sel(time=pulse_midpoint, method="nearest"), 
            s=100, edgecolor='white', label='Buoy Locations', 
            cmap = "coolwarm", vmin=vmin, vmax=vmax)

# Set plot labels and title
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title(f'Interpolated Spatial Anomaly at Pulse Midpoint (Day {pulse_midpoint})')

plt.legend()
plt.show()
```

```python
# Calculate the true anomaly field at the pulse midpoint (from the full field minus the mean)
true_anomaly_field = space_time_grid.sel(time=pulse_midpoint) - space_time_grid.mean(dim='time')

# Predict the reconstructed anomaly field at the pulse midpoint using GPR (from previous code)
y_grid_pred_reshaped = y_grid_pred.reshape(grid_size, grid_size)

# Plot the comparison between the reconstructed field and the true anomaly field
plt.figure(figsize=(12, 6))

# --------------------------
# Subplot 1: True Anomaly Field
# --------------------------
plt.subplot(1, 2, 1)
plt.pcolormesh(true_anomaly_field.lat, true_anomaly_field.lon, 
               true_anomaly_field.values, 
               cmap='coolwarm', vmin=vmin, vmax=vmax)
plt.colorbar(label="Temperature Anomaly (True)")
plt.scatter(random_buoy_locs[:, 0], random_buoy_locs[:, 1], c='black', s=100, edgecolor='white', label='Buoy Locations')
plt.title(f'True Anomaly Field at Pulse Midpoint (Day {pulse_midpoint})')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

# --------------------------
# Subplot 2: GPR Reconstructed Anomaly Field
# --------------------------
plt.subplot(1, 2, 2)
plt.pcolormesh(grid_lat, grid_lon, y_grid_pred_reshaped, 
               cmap='coolwarm', vmin=vmin, vmax=vmax)
plt.colorbar(label="Temperature Anomaly (Reconstructed)")
plt.scatter(random_buoy_locs[:, 0], random_buoy_locs[:, 1], c='black', s=100, edgecolor='white', label='Buoy Locations')
plt.title(f'GPR Reconstructed Anomaly Field at Pulse Midpoint (Day {pulse_midpoint})')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

# Adjust layout for better display
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

# ------------------------------
# Function 1: Generate Random Buoy Locations
# ------------------------------
def generate_random_buoy_locations(n_buoys, lat_range, lon_range):
    """
    Generates random buoy locations within specified latitude and longitude ranges.
    
    Parameters:
        n_buoys (int): Number of buoys to generate.
        lat_range (tuple): Latitude range (min, max).
        lon_range (tuple): Longitude range (min, max).
    
    Returns:
        np.ndarray: Array of random buoy locations (lat, lon) with shape (n_buoys, 2).
    """
    buoy_locs = np.column_stack((
        np.random.uniform(low=lat_range[0], high=lat_range[1], size=n_buoys),
        np.random.uniform(low=lon_range[0], high=lon_range[1], size=n_buoys)
    ))
    return buoy_locs

# ------------------------------
# Function 2: Extract Buoy Data from the Spatiotemporal Field
# ------------------------------
def extract_buoy_data(space_time_grid, buoy_locs, time_feature):
    """
    Extracts buoy data from the spatiotemporal field based on the buoy locations.
    
    Parameters:
        space_time_grid (xr.DataArray): The spatiotemporal field (e.g., synthetic field).
        buoy_locs (np.ndarray): Array of buoy locations (lat, lon) with shape (n_buoys, 2).
        time_feature (np.ndarray): Time feature (array of time points).
    
    Returns:
        xr.DataArray: Buoy data with shape (n_days, n_buoys).
    """
    n_days = len(time_feature)
    n_buoys = len(buoy_locs)
    
    # Initialize the buoy data array
    buoy_data = xr.DataArray(
        np.zeros((n_days, n_buoys)),
        coords=[time_feature, np.arange(n_buoys)],
        dims=["time", "buoy"]
    )
    
    # Extract the data for each buoy based on the nearest lat/lon values
    for b, (lat, lon) in enumerate(buoy_locs):
        buoy_data[:, b] = space_time_grid.sel(lat=lat, lon=lon, method="nearest").values
    
    return buoy_data

# ------------------------------
# Function 3: Add Noise to Buoy Data
# ------------------------------
def add_noise_to_buoy_data(buoy_data, noise_scale):
    """
    Adds Gaussian noise to the buoy data.
    
    Parameters:
        buoy_data (xr.DataArray): Buoy data (n_days, n_buoys).
        noise_scale (float): The standard deviation of the noise to add.
    
    Returns:
        xr.DataArray: Noisy buoy data.
    """
    noisy_buoy_data = buoy_data.copy()  # Copy to avoid modifying the original data
    noisy_buoy_data += np.random.normal(0, noise_scale, size=noisy_buoy_data.shape)
    return noisy_buoy_data

# ------------------------------
# Function 4: Plot Buoy Data (Optional Utility)
# ------------------------------
def plot_buoy_data(buoy_data, buoy_locs, title="Buoy Data Over Time"):
    """
    Plots buoy data over time for each buoy.
    
    Parameters:
        buoy_data (xr.DataArray): Buoy data (n_days, n_buoys).
        buoy_locs (np.ndarray): Array of buoy locations (lat, lon) with shape (n_buoys, 2).
        title (str): Title of the plot.
    """
    plt.figure(figsize=(10, 6))
    for b in range(buoy_data.shape[1]):
        plt.plot(buoy_data.time, buoy_data[:, b], label=f"Buoy {b} (Lat: {buoy_locs[b, 0]:.2f}, Lon: {buoy_locs[b, 1]:.2f})")
    plt.title(title)
    plt.xlabel("Time (Days)")
    plt.ylabel("Temperature")
    plt.legend()
    plt.show()

# ------------------------------
# Example Usage: Putting It All Together
# ------------------------------

# Parameters
n_buoys = 5
lat_range = (-10, 10)
lon_range = (-10, 10)
n_days = 200
time_feature = np.arange(1, n_days + 1)
noise_scale = 0.5

# Step 1: Generate Random Buoy Locations
random_buoy_locs = generate_random_buoy_locations(n_buoys, lat_range, lon_range)

# Step 2: Assume you already have the 'space_time_grid' (spatiotemporal field)
# We would use the function to extract buoy data from this field
# Example: space_time_grid = xr.DataArray(...)

# Step 3: Extract Buoy Data from the Spatiotemporal Field
buoy_data = extract_buoy_data(space_time_grid, random_buoy_locs, time_feature)

# Step 4: Add Noise to the Buoy Data
noisy_buoy_data = add_noise_to_buoy_data(buoy_data, noise_scale)

# Step 5: Plot the Noisy Buoy Data (optional)
plot_buoy_data(noisy_buoy_data, random_buoy_locs, title="Noisy Buoy Data Over Time")
```
